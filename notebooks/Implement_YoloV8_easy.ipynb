```python
# split the list into train and save some of the annotations for a validation set (going from the end of the previous list)

def split_list(original_list, val_size):
    """
    Splits a list into two parts: one with the last 8 entries, and the other with the rest.

    :param original_list: The list to be split.
    :return: A tuple of two lists: (rest_of_list, last_eight_entries)
    """
    # Ensure the list has more than val elements
    if len(original_list) <= val_size:
        return [], original_list

    # Split the list
    last_val_entries = original_list[-val_size:]  # Gets the last 8 entries
    rest_of_list = original_list[:-val_size]  # Gets the rest of the list

    return rest_of_list, last_val_entries
```

```python
# copy the images to a specific directory
def copy_images_to_directory(image_names, destination_directory, original_location):
    """
    Copies a list of images to a specified destination directory.

    :param image_paths: List of strings, where each string is the path to an image file.
    :param destination_directory: String, path to the destination directory where images will be copied.
    """
    # Ensure the destination directory exists, create it if it doesn't
    os.makedirs(destination_directory, exist_ok=True)

    # Loop through each image path in the list
    for image_name in image_names:
        # Determine the name of the image file
        image_name = os.path.basename(image_name)

        # Construct the full path for the root of the image
        original_path = os.path.join(original_location, image_name)

        # Construct the full path for the destination of the image
        destination_path = os.path.join(destination_directory, image_name)

        # Copy the image to the destination directory
        shutil.copy(original_path, destination_path)
        print(f"Copied {image_name} to {destination_path}")
```

```python
# copy images across to train
destination_directory_train = f'{DATASET_DIRECTORY}/images/train'

original_location = f'{ROOT_DIRECTORY}/training_frames'

copy_images_to_directory(train_list, destination_directory_train, original_location)
```

```python
def create_subseries_from_pandas(original_series, index_list):
    """
    Creates a new pandas Series containing only the entries that match the index names
    in the given list.

    :param original_series: The original pandas Series.
    :param index_list: A list of index names to include in the new Series.
    :return: A new pandas Series with entries for the specified index names.
    """
    new_series = original_series.loc[index_list]
    return new_series
```

```python
new_series = create_subseries_from_pandas(series, val_list)
```

```python
# create text files for the all bounding boxes in a series
def create_bbox_files(series, destination_directory):
    """
    Creates a text file for each entry in the series, with lines for each bounding box.

    :param series: A dictionary with the described structure.
    """
    for file_name, data in series.items():
        # Define the text file name, appending '.txt' to the index (file name)
        text_file_name = f"{destination_directory}/{file_name}.txt"

        # Open the text file in write mode
        with open(text_file_name, 'w') as file:
            # Iterate through each bounding box in the bounding boxes list
            for bbox in data['bboxes']:
                # Write the bounding box to the file in the specified format
                bbox_line = '0 ' + ' '.join(map(str, bbox)) + '\n'
                file.write(bbox_line)

            print(f"Created {text_file_name}")
```

```python
# remove .jpg from the label names
def rename_files(directory):
    """
    Renames all files in the specified directory that end with '.jpg.txt' to remove the '.jpg' part.

    :param directory: The directory containing the files to be renamed.
    """
    for filename in os.listdir(directory):
        # Check if the file ends with '.jpg.txt'
        if filename.endswith('.jpg.txt'):
            # Construct the new filename by removing '.jpg'
            new_filename = filename.replace('.jpg', '')

            # Construct the full file paths
            old_file_path = os.path.join(directory, filename)
            new_file_path = os.path.join(directory, new_filename)

            # Rename the file
            os.rename(old_file_path, new_file_path)
            print(f"Renamed '{filename}' to '{new_filename}'")
```

```python
directory = f'{DATASET_DIRECTORY}/labels/train'  # Replace with the path to your directory
rename_files(directory)
```
